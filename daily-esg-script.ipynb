```python
df=pd.read_csv(r'Datasets/esg_texts.csv').sort_values('Date',ascending=False).reset_index(drop=True)
```

```python
df['text']=df['text'].str.strip()
```

```python
df['text']=df['text'].apply(lambda x:x.replace("Is ESG investing about reducing carbon emissions or other goals? That’s among the topics covered in the Bloomberg ESG survey. It’s not long, and we don’t collect your name or any contact information. Please click hereto share your views.",'').strip())
df['text']=df['text'].apply(lambda x:x.replace("Is ESG a fad or an investment approach that will last? That’s among the topics covered in the Bloomberg ESG survey. It’s not long, and we don’t collect your name or any contact information. Click hereto share your views.",'').strip())
df['text']=df['text'].apply(lambda x:x.replace("Is ESG a fad or an investment approach that will last? That’s among the topics covered in the Bloomberg ESG survey. It’s not long, and we don’t collect your name or any contact information. Please click hereto share your views.",'').strip())
```

```python
df['headline']=df['headline'].apply(lambda x:x.replace('[EXTERNAL] ESG DAILY:','').strip())
df['text']=df['text'].apply(lambda x:x.replace('Alert: ESG DAILY <https://blinks.bloomberg.com/news/topics/ESG%20DAILY','').strip())
df['text']=df['text'].apply(lambda x:x.replace('Bloomberg','').strip())
df['text']=df['text'].apply(lambda x:x.split('ESG FIXTURES')[0])
df['text']=df['text'].apply(lambda x:x.split('Share your views')[0])
df['headline']=df.headline.str.replace("���","'").str.replace("��","'")
df['Remove']=df.text.str.contains('Alert:')

df=df[df.Remove==False].reset_index(drop=True).drop('Remove',axis=1)
```
